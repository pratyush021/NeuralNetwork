```
import numpy as np 
import matplotlib.pyplot as plt
```

```
n_pts = 500
top = np.array([np.random.normal(10,2,n_pts),np.random.normal(10,2,n_pts)]).T
bot = np.array([np.random.normal(5,2,n_pts),np.random.normal(5,2,n_pts)]).T
```

```
xtr = np.vstack((top,bot))
ytr = np.matrix(np.append(np.ones(n_pts),
                          np.zeros(n_pts))).T
```

```

plt.scatter(xtr[:n_pts,0],xtr[:n_pts,1])
plt.scatter(xtr[n_pts:,0],xtr[n_pts:,1]) 
plt.show()
```

```
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import adam
```

```
#Sequential = linear stack of layers
#Dense      = Everynode of previous layer is connected to everynode of next layer 
#Adam       = Optimization function that controls learning rate Recomended by keras 
```

```
model = Sequential()
model.add(Dense(units = 1 , input_dim = 2, activation='sigmoid'))
model.compile(adam(lr = 0.1),loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

```
print(model.summary())
```

```
h = model.fit(xtr,ytr,epochs = 10, verbose = 1)
#verbose = 1 shows the process and information of accuracy 
```

```
plt.plot(h.history['loss'])
plt.show()
```

```
plt.plot(h.history['acc'])
plt.show()
```

```
model.predict_classes(np.array([[2,8]]))
```
